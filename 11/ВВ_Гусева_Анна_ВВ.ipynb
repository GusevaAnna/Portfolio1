```python
# Фильтруем задачи типа "Инцидент - БАГ"
bugs_df = df[df['task_type'].str.contains('Инцидент - БАГ', case=False, na=False)]
```

```python
# Общее количество багов
total_bugs = bugs_df.shape[0]
print(f"Общее количество багов: {total_bugs}")

bugs_df['date_creation'] = pd.to_datetime(bugs_df['date_creation'])

# Установка 'date_creation' в качестве индекса
bugs_df.set_index('date_creation', inplace=True)

# Динамика по месяцам
monthly_bugs = bugs_df.resample('M').size()

# Сортировка данных по убыванию
monthly_bugs = monthly_bugs.sort_index(ascending=False)

# Визуализация
fig, ax = plt.subplots(figsize=(12, 6))

# Построение горизонтальной столбчатой диаграммы
monthly_bugs.plot(kind='barh', ax=ax)

# Форматирование меток оси Y
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: monthly_bugs.index[x].strftime('%Y-%m')))

# Подписи осей
plt.title('Динамика количества багов по месяцам')
plt.xlabel('Количество багов')
plt.ylabel('Месяц')

# Поворот меток для улучшения читаемости
plt.tight_layout()
plt.show()
```

# Группировка и подсчет количества багов по приоритетам для каждой команды

```python
bugs_df['priority_general'].value_counts()
```

```python
# Удаление строк с пропущенными значениями в столбце 'priority_general'
bugs_df = bugs_df.dropna(subset=['priority_general'])

# Группировка данных по команде и приоритету
priority_counts = bugs_df.groupby(['executor_team', 'priority_general']).size().unstack(fill_value=0)

# Добавляем колонку с общим количеством багов по команде для сортировки
priority_counts['total'] = priority_counts.sum(axis=1)

# Преобразуем данные в числовой формат (если необходимо) и убираем пропущенные значения
priority_counts['2. Высокий'] = pd.to_numeric(priority_counts['2. Высокий'], errors='coerce').fillna(0)

# Сортировка по количеству багов с высоким приоритетом (предполагается, что '2. Высокий' является высоким приоритетом)
priority_counts_sorted = priority_counts.sort_values(by='2. Высокий')

# Визуализация
fig, ax = plt.subplots(figsize=(15, 15))
priority_counts_sorted.drop(columns='total').plot(kind='barh', stacked=True, ax=ax, colormap='viridis')

plt.title('Количество багов по приоритетам для каждой команды (сортировка по высокому приоритету)')
plt.xlabel('Количество багов')
plt.ylabel('Команда')
plt.legend(title='Приоритет')
plt.tight_layout()
plt.show()
```

# Среднее время решения багов по командам

```python
# Среднее время решения багов по командам
avg_resolution_time_by_team = bugs_df.groupby('executor_team')['resolution_time'].mean()
avg_resolution_time_by_team_sorted = avg_resolution_time_by_team.sort_values(ascending=False)
print("\nСреднее время решения багов по командам:")
avg_resolution_time_by_team_sorted.head(50)
```

```python
# Визуализация
fig, ax = plt.subplots(figsize=(15, 15))

# Построение горизонтальной столбчатой диаграммы
avg_resolution_time_by_team.sort_values().plot(kind='barh', ax=ax, color='skyblue')

plt.title('Среднее время решения багов по командам')
plt.xlabel('Среднее время решения (дни)')
plt.ylabel('Команда')
plt.tight_layout()

# Показать график
plt.show()
```
